```python
import pandas as pd

# Example data
n_data2 = n_data

n_data2['Oral Cancer (Diagnosis)'] = [1 if x == 'Yes' else 0 for x in n_data2['Oral Cancer (Diagnosis)']]

```

```python
x = n_data.drop(columns=['Oral Cancer (Diagnosis)']
                )
y = n_data['Oral Cancer (Diagnosis)']
```

Encoding Process

```python
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)
```

```python
from category_encoders import BinaryEncoder
import pandas as pd

categorical_features = ['Country', 'Gender', 'Tobacco Use', 'Alcohol Consumption',
       'HPV Infection', 'Betel Quid Use', 'Chronic Sun Exposure',
       'Poor Oral Hygiene', 'Diet (Fruits & Vegetables Intake)',
       'Family History of Cancer', 'Compromised Immune System', 'Oral Lesions',
       'Unexplained Bleeding', 'Difficulty Swallowing',
       'White or Red Patches in Mouth', 'Treatment Type', 'Early Diagnosis',]
      
encoder = BinaryEncoder()
x_encoded = encoder.fit_transform(x[categorical_features])
x = x.drop(categorical_features, axis=1)
x = x.reset_index(drop=True)
x_encoded = x_encoded.reset_index(drop=True)
x = pd.concat([x, x_encoded], axis=1)

# 'Age','Tumor Size (cm)','Cancer Stage','Survival Rate (5-Year, %)','Cost of Treatment (USD)', 'Economic Burden (Lost Workdays per Year)',
#                         'Country_0','Country_1','Country_2','Country_3','Country_4','Gender_0','Gender_1','Tobacco Use_0','Tobacco Use_1','Alcohol Consumption_1',
#                         'HPV Infection_0','HPV Infection_0','Betel Quid Use_0','Betel Quid Use_0','Gender_0'
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dcs = DecisionTreeClassifier()
dcs.fit(x_train, y_train)
```

```python
prediction = dcs.predict(x_test)
```

```python
from sklearn.metrics import(
    accuracy_score,
    precision_score
)
prediction = dcs.predict(x_test)
accuracy = accuracy_score (prediction,y_test)
precision = precision_score (prediction, y_test)


print(f"accuracy: {accuracy*100}")
print(f"precision: {precision*100}")
```

```python
n_prediction = dcs.predict(x_test.iloc[[70]])
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x, y = smote.fit_resample (x,y)
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(x,y,test_size=.25,random_state=42)
```

```python
dcs1 = DecisionTreeClassifier()
dcs1.fit(x_train1, y_train1)
```

```python
from sklearn.metrics import accuracy_score, precision_score

dcs_prediction = dcs1.predict(x_test1)

dcs_accuracy = accuracy_score(y_test1, dcs_prediction)
dcs_precision = precision_score(y_test1, dcs_prediction)

print(f"accuracy: {dcs_accuracy*100}")
print(f"precision: {dcs_precision*100}")
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import graphviz
from joblib import dump, load
import joblib

dsc1 = DecisionTreeClassifier()
dsc1.fit(x_train1, y_train1)

export_graphviz(dsc1, out_file="tree.dot", filled=True, feature_names=x_train1.columns)

with open("tree.dot") as f:
    dot_graph = f.read()
    graph = graphviz.Source(dot_graph)
    graph.view()  

dump(dsc1, "decision_tree_model.joblib")
```
